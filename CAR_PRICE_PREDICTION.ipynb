```python
### Make categories based on average price
cars=df.copy()
cars.groupby(['Company Name'])['price'].mean().reset_index()
flag=cars.merge(cars.groupby(['Company Name'])['price'].mean().reset_index(),how='left',on="Company Name")
table=pd.cut(flag['price_y'],bins=[0,10000,20000,40000],labels=['Budget','Medium','Luxury'],right=False)
df['range']=table
df.head()
```

```python
from ctypes import DEFAULT_MODE
### Convert Catregories into Numerical for regression
# Defining the map function
def convert_label(label,data):
    df = pd.get_dummies(data[label])
    data = pd.concat([data, df], axis = 1)
    data=data.drop([label], axis = 1)
    return data
df = convert_label('fueltype',df)
df = convert_label('carbody',df)
df = convert_label('enginetype',df)
df = convert_label('aspiration',df)
df = convert_label('drivewheel',df)
df = convert_label('enginelocation',df)
df = convert_label('fuelsystem',df)
df = convert_label('range',df)
df = convert_label('Company Name',df)
```

#**Train-Test Split and feature scaling**

```python
df.columns
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
header = ['wheelbase', 'curbweight','enginesize','horsepower','citympg','highwaympg','carlength','carwidth','price']
df[header] = scaler.fit_transform(df[header])
df.head()
```

```python
labels=pd.DataFrame({'price':df['price'].to_list()})
labels.head()
```

```python
dataset=df.drop(['price','carheight','CarName','car_ID'],axis=1)
dataset.head()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test  = train_test_split(dataset,
                                              labels,
                                              test_size=0.2,
                                              random_state = 1)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
y_pred
```

```python
from sklearn.metrics import mean_squared_error,r2_score,mean_absolute_percentage_error
mean_squared_error(y_test, y_pred)
```

```python
print("R2 SCORE is ",r2_score(y_test,y_pred))
```

```python
print("MAPE is ",mean_absolute_percentage_error(y_test,y_pred))
```
